```python
from nltk.tokenize import word_tokenize, sent_tokenize
from collections import Counter
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python

SOS = '<s>'
EOS = '</s>'

# Define or load input
document = '''
He blinked. It wasn't the Snitch at all. It was a pair of glasses. How strange.

He blinked again. The smiling face of Albus Dumbledore swam into view above him.

"Good afternoon, Harry," said Dumbledore. Harry stared at him. Then he remembered: "Sir! The Stone! It was Quirrell! He's got the Stone! Sir, quick --"

"Calm yourself, dear boy, you are a little behind the times," said Dumbledore. "Quirrell does not have the Stone."

"Then who does? Sir, I --"

"Harry, please relax, or Madam Pomfrey will have me thrown out.

Harry swallowed and looked around him. He realized he must be in the hospital wing. He was lying in a bed with white linen sheets, and next to him was a table piled high with what looked like half the candy shop.

"Tokens from your friends and admirers," said Dumbledore, beaming. "What happened down in the dungeons between you and Professor Quirrell is a complete secret, so, naturally, the whole school knows. I believe your friends Misters Fred and George Weasley were responsible for trying to send you a toilet seat. No doubt they thought it would amuse you. Madam Pomfrey, however, felt it might not be very hygienic, and confiscated it."
'''

fo = open('Harry Potter Book 1.txt', 'r', encoding = 'utf-8')
document = fo.read()

print()
print('Input, document')
print(document[:500])

```

```python
# Split into sentences
sentences = sent_tokenize(document)

# Tokenize a list of sentences
tokens = [word_tokenize(sent) for sent in sentences]

# Add start of sentence tokens
tokens = [[SOS] + sent for sent in tokens]

# Add end of sentence tokens
tokens = [sent + [EOS] for sent in tokens]

# Make lowercase
tokens = [[token.lower() for token in sent] for sent in tokens]

for i, sent in enumerate(tokens[:5]):
    print(i, sent)
```

```python
# Get unigram counts
unigram_counts = Counter()
for sent in tokens:
    for token in sent:
        unigram_counts[token] += 1

print(f'Unigram counts:\n{unigram_counts}')
```

```python
# Get bigram counts
bigram_counts = Counter()
for sent in tokens:

    # sent      [w1, w2, w3, ... , w(n-1), w(n)]

    # previous  [w1, w2, w3, ... , w(n-1)]
    previous = sent[:-1]

    # current   [w2, w3, ... , w(n-1), w(n)]
    current = sent[1:]

    assert len(previous) == len(current)

    for p, c in zip(previous, current):
        bigram_counts[(p, c)] += 1

print(f'Bigram counts:\n{bigram_counts}')
```

```python
# Iterating over counter
for (previous, current), count in bigram_counts.items():
    print(previous, current, count)
```

```python
# As dataframe
# Iterating over counter
rows = []
for (previous, current), count in bigram_counts.items():
    rows.append((previous, current, count))
df = pd.DataFrame(rows, columns=['wi-1', 'wi', 'count'])
print(df)

df.to_csv('output.csv', encoding = 'utf-8-sig')
```
